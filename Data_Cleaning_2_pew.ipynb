```python
weather = pd.read_csv('/content/weather.csv')
```

```python
weather
```

```python
weather_melt = (pd.melt(weather, id_vars=
                        ['id','year','month', 'element'],var_name='day',
                        value_name= 'temp'))
```

```python
weather_melt.head()
```

```python
# We use the pivot table method to creat a column from the values of element column
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month', 'day'],
    columns='element',
    values= 'temp'
    
)
```

```python
weather_tidy
```

```python
# reset index is use to reset hierachy index
weather_tidy.reset_index()
```
